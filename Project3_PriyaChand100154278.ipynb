```python
high_rent_rate_areas = BC_Census[BC_Census['shelt_rent_30plus_rate'] > 50]
print(high_rent_rate_areas[['chsa', 'shelt_rent_30plus_rate']])
```

```python
# Analyze household income and rent prices in high rent rate areas
income_rent_analysis = high_rent_rate_areas[['chsa', 'income_total_mean', 'shelt_rent_mo_cost_mean']]
print(income_rent_analysis)
```

```python
# Display CHSAs and their relevant columns
print(high_rent_rate_areas[['chsa', 'shelt_rent_30plus_rate', 'income_total_mean', 'shelt_rent_mo_cost_mean']])

# Analyze household income and rent prices in high rent rate areas
income_rent_analysis = high_rent_rate_areas[['chsa', 'income_total_mean', 'shelt_rent_mo_cost_mean']]
print(income_rent_analysis)
```

```python
import pandas as pd
import numpy as np

# Calculate the percentage of income spent on rent
income_rent_analysis['percent_income_spent_on_rent'] = (income_rent_analysis['shelt_rent_mo_cost_mean'] / income_rent_analysis['income_total_mean']) * 100

# Display the updated analysis
print(income_rent_analysis[['chsa', 'income_total_mean', 'shelt_rent_mo_cost_mean', 'percent_income_spent_on_rent']])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Draw a bar graph for the percentage of income spent on rent
plt.figure(figsize=(10, 6))
sns.barplot(x='chsa', y='percent_income_spent_on_rent', data=income_rent_analysis, color='lightgreen', edgecolor='black')
plt.title('Bar Graph of Percentage of Income Spent on Rent')
plt.xlabel('CHSA')
plt.ylabel('Percentage of Income Spent on Rent')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```

```python
# Examine data from regional level
regional_data = BC_Census[['pha', 'shelt_rent_mo_cost_mean', 'income_total_mean']]
```

```python
# Group data by PHA and calculate average subsidization rate, average income, and average rent
regional_summary = regional_data.groupby('pha').agg({
    'shelt_rent_mo_cost_mean': 'mean',
    'income_total_mean': 'mean'
}).reset_index()
```

```python
# Calculate the average subsidization rate for each region
regional_summary['avg_subsidization_rate'] = (regional_summary['shelt_rent_mo_cost_mean'] / regional_summary['income_total_mean']) * 100
```

```python
# Display the summary
print(regional_summary[['pha', 'avg_subsidization_rate']])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


# Draw a bar graph for visual comparison
plt.figure(figsize=(10, 6))
sns.barplot(x='pha', y='avg_subsidization_rate', data=regional_summary, color='lightpink', edgecolor='black')
plt.title('Average Subsidization Rates Across PHA Regions')
plt.xlabel('PHA Regions')
plt.ylabel('Average Subsidization Rate')
plt.show()
```
